```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['Surname']=label_encoder.fit_transform(df['Surname'])
df['Geography']=label_encoder.fit_transform(df['Geography'])
df['Gender']=label_encoder.fit_transform(df['Gender'])
print(df.info())
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!='Exited']
y = df.iloc[:,df.columns=='Exited']
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
xtrain.head()
```

```python
xtrain.shape
```

```python
ytrain.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output=model.predict(xtest)
print(predict_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score for RF:',acc)
```
