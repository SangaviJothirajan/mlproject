```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
a['trans_date_trans_time']=label_encoder.fit_transform(a['trans_date_trans_time'])
a['merchant']=label_encoder.fit_transform(a['merchant'])
a['category']=label_encoder.fit_transform(a['category'])
a['first']=label_encoder.fit_transform(a['first'])
a['last']=label_encoder.fit_transform(a['last'])
a['gender']=label_encoder.fit_transform(a['gender'])
a['street']=label_encoder.fit_transform(a['street'])
a['city']=label_encoder.fit_transform(a['city'])
a['state']=label_encoder.fit_transform(a['state'])
a['job']=label_encoder.fit_transform(a['job'])
a['dob']=label_encoder.fit_transform(a['dob'])
a['trans_num']=label_encoder.fit_transform(a['trans_num'])
print(a.info())
```

```python
a.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = a.iloc[:,a.columns!='is_fraud']
y = a.iloc[:,a.columns=='is_fraud']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
xtrain.shape
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output=model.predict(xtest)
print(predict_output)
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score for RF:',acc)
```
