```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['v1']=label_encoder.fit_transform(df['v1'])
df['v2']=label_encoder.fit_transform(df['v2'])
df['Unnamed: 2']=label_encoder.fit_transform(df['Unnamed: 2'])
df['Unnamed: 3']=label_encoder.fit_transform(df['Unnamed: 3'])
df['Unnamed: 4']=label_encoder.fit_transform(df['Unnamed: 4'])
print(df.info())
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!='v1']
y = df.iloc[:,df.columns=='v1']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
xtrain.shape
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output=model.predict(xtest)
print(predict_output)
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score for RF:',acc)
```
